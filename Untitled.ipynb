# COVID Analysis of Relationship Between Tests and Positive Test Rates

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.gridspec import GridSpec
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

from pathlib import Path
import numpy as np
import pandas as pd
import datetime
import time

from src.phac import PHAC

can = PHAC()
```

```python
fit_win = 14
meas_win = 7
tests = can.var_by_location("new_tests_per_thousand", *can.provinces, ma_window=1).fillna(0)
cases = can.var_by_location("new_cases_per_million", *can.provinces, ma_window=1)
deaths = can.var_by_location("new_deaths_per_million", *can.provinces, ma_window=1).fillna(0)
pt_rate = can.pos_test_rate(1, *can.provinces)
pt_growth = can.growth_rate("pos_test_rate", meas_win, *can.provinces)
```

```python
tests
```

```python
tests_adj = tests - can.var_by_location("new_tests_per_thousand", *can.provinces, ma_window=meas_win)
tests_adj["Ontario"].plot(figsize=(16,12))
```

```python
pt_rate_adj = pt_rate - can.pos_test_rate(meas_win, *can.provinces)
pt_rate_adj["Ontario"].plot(figsize=(16,12))
```

```python
location = "Ontario"
pt_rate_adj_loc = pt_rate_adj[location].values
valid_vals = np.nonzero(
    np.logical_not(np.logical_or(
        np.isnan(pt_rate_adj_loc), np.isinf(pt_rate_adj_loc)
    ))
)
pt_rate_adj_loc = pt_rate_adj_loc[valid_vals]
tests_adj_loc = (tests_adj[location].values)[valid_vals]
print(len(pt_rate_adj_loc), len(tests_adj_loc))
```

```python
np.correlate(tests_adj_loc, pt_rate_adj_loc)
```
